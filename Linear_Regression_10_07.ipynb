```python
#Linear regression
%matplotlib inline
import numpy as np
from pylab import *
```

```python
pageTime = np.random.normal(3.0,1.0,1000)
pageDollar = 100+(pageTime + np.random.normal(0,0.1,1000))*3
scatter(pageTime,pageDollar)
```

```python
from scipy import stats
slope,intercept,r_value,p_value,std_dev = stats.linregress(pageTime,pageDollar)
```

```python
import matplotlib.pyplot as plt
def predict(x):
    return slope*x+intercept

fitline = predict(pageTime)
plt.scatter(pageTime,pageDollar)
plt.plot(pageTime,fitline,c="r")
plt.show()
```
